```python
import requests
import time
import sys

# ======================================================
# Step 1: Configuration
# ======================================================

#API_KEY = "Removed my API key and regenerated it as well"
BASE_URL = "https://www.presentmeapp.xyz"

CREATE_ENDPOINT = f"{BASE_URL}/api/v1/presentations"
STATUS_ENDPOINT = f"{BASE_URL}/api/v1/presentations/status"
LIST_ENDPOINT = f"{BASE_URL}/api/v1/presentations"

MAX_ATTEMPTS = 120
WAIT_SECONDS = 10

headers = {
    "Authorization": f"Bearer {API_KEY}"
}

# ======================================================
# Step 2: Build multipart/form-data payload
# ======================================================

topic_description = (
    "Create a 10-slide investor presentation about AI Agents in Production. "
    "Include: a title slide with heading 'AI Agents in Production From MCP to Market'; "
    "a problem statement slide about why most AI projects fail before reaching production; "
    "a slide defining AI agents and comparing them to chatbots and RPA; "
    "a slide explaining the Model Context Protocol as the standard for connecting agents to enterprise tools; "
    "a slide with real-world use cases in customer support, data analysis, and financial reporting; "
    "a slide on responsible AI covering fairness, transparency, and accountability; "
    "a slide on testing strategies for AI agents; "
    "a slide on production deployment best practices; "
    "a slide on ROI with cost reduction benchmarks; "
    "and a closing slide with a 90-day roadmap and call to action."
)

payload = {
    "topic": (None, topic_description),
    "slide_count": (None, "10"),
    "creation_mode": (None, "thematic"),
    "audience": (None, "investors and C-suite executives")
}

# ======================================================
# Step 3: Create presentation (async)
# ======================================================

try:
    response = requests.post(
        CREATE_ENDPOINT,
        headers=headers,
        files=payload,
        timeout=30
    )

    if response.status_code == 402:
        print("Insufficient credits. Please top up your account.")
        sys.exit(1)

    response.raise_for_status()
    data = response.json()

    task_id = data.get("task_id")
    if not task_id:
        print("Error: task_id not returned by API.")
        sys.exit(1)

except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
    sys.exit(1)

# ======================================================
# Step 4: Poll generation status
# ======================================================

for attempt in range(1, MAX_ATTEMPTS + 1):
    try:
        status_response = requests.get(
            f"{STATUS_ENDPOINT}/{task_id}",
            headers=headers,
            timeout=15
        )
        status_response.raise_for_status()

        status_data = status_response.json()
        status = status_data.get("status", "UNKNOWN")

        print(f"[Attempt {attempt}/{MAX_ATTEMPTS}] Status: {status}... waiting {WAIT_SECONDS}s")

        if status == "SUCCESS":
            break

        if status == "FAILURE":
            print("Presentation generation failed.")
            sys.exit(1)

        time.sleep(WAIT_SECONDS)

    except requests.exceptions.RequestException as e:
        print(f"Status polling error: {e}")
        sys.exit(1)

else:
    print("========================================")
    print("Generation still in progress.")
    print(f"Task ID: {task_id}")
    print("Please check your PresentMe dashboard.")
    print("========================================")
    sys.exit(0)

# ======================================================
# Step 5: Retrieve presentation details
# ======================================================

try:
    list_response = requests.get(
        LIST_ENDPOINT,
        headers=headers,
        timeout=30
    )
    list_response.raise_for_status()

    list_data = list_response.json()
    presentations = list_data.get("presentations", [])

    if not presentations:
        print("No presentations found.")
        sys.exit(1)

    presentation_id = presentations[0].get("id")

    details_response = requests.get(
        f"{LIST_ENDPOINT}/{presentation_id}",
        headers=headers,
        timeout=30
    )
    details_response.raise_for_status()

    presentation = details_response.json()
    title = presentation.get("title", "Untitled")
    slides = presentation.get("slides", [])
    slide_count = len(slides) if isinstance(slides, list) else "Unknown"

except requests.exceptions.RequestException as e:
    print(f"Error retrieving presentation: {e}")
    sys.exit(1)

# ======================================================
# Final Output (EXACT FORMAT)
# ======================================================

print("========================================")
print("Presentation Generated Successfully!")
print("========================================")
print(f"Title: {title}")
print(f"Slides: {slide_count}")
print(f"Presentation ID: {presentation_id}")
print("========================================")
```
